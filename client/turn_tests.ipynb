```python
import time
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
TURN_ANGLE = 90
TURN_STEPS = 10
```

```python
steering_increment = TURN_ANGLE/TURN_STEPS
current_angle = 0
current_angle_array = []
step_array = []
for i in range(TURN_STEPS):
    current_angle += steering_increment
    step_array.append(i)
    current_angle_array.append(current_angle)
```

```python
step_array
```

```python
current_angle_array
```

```python
plt.figure()
plt.plot(step_array,current_angle_array, label = "Angle at different steps")
plt.xlabel("Step")
plt.ylabel("Angle")
plt.show()
```
